## Questions

1) What is the average price of cars across different brands, and which brands have the highest and lowest average prices.
2) How does the condition of a car (Status) affect its price.
3) Is there a correlation between mileage (Kilometers) and the price of a car.
4) How does engine capacity (CC) relate to car price and mileage

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt              
import seaborn as sns
```

```python
Car = pd.read_csv('cars_info.csv')
```

## What is the average price of cars across different brands, and which brands have the highest and lowest average prices

```python
Car.head(1)
```

```python
Brand = Car.groupby("Brand")["Price"].mean().reset_index()
```

```python
Brand
```

```python
plt.figure(figsize = (6,10))
sns.barplot(data = Brand,x="Price",y="Brand",hue = "Price")
```

```python
Brand = Car.groupby("Brand")["Price"].mean().reset_index()
sort = Brand.sort_values(by='Price', ascending=False)
```

```python
sort
```

```python
h = sort.iloc[0]
l = sort.iloc[-1]
```

```python
h
```

```python
l
```

## How does the condition of a car (Status) affect its price.

```python
c=Car[["Name","Status","Price"]].value_counts().reset_index()
```

```python
c
```

```python
sns.pairplot(c[['Status', 'Price','count']],hue = "Status")
```

```python
sns.barplot(data = c,x="Status",y="Price",hue = "Price")
```

```python
sns.relplot(data = c,x="Status",y="Price",hue = "Price")
```

## Is there a correlation between mileage (Kilometers) and the price of a car.

```python
Car.head(1)
```

```python
C = Car[['Kilometers','Price']].corr()
```

```python
C
```

```python
sns.barplot(data = C,x="Kilometers",y="Price")
```

```python
sns.heatmap(C,annot = True)
```

## How does engine capacity (CC) relate to car price and mileage

```python
cc = Car[["CC","Kilometers","Price"]].value_counts().reset_index()
```

```python
cc
```

```python
sns.barplot(data = cc,x="Price",y="CC")
```

```python
sns.heatmap(cc,annot=True)
```

```python
sns.boxenplot(data = cc,x="Price",y="CC")
```

```python
sns.scatterplot(data = cc,x="CC",y="Kilometers",hue="Price")
```
